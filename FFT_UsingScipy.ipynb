```python
from scipy.fftpack import fft
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 600 #number of sample points
```

```python
dt = 1./800 #sample spacing
```

```python
x = np.linspace(0, N*dt, N)
y = np.sin(50*2*np.pi*x) + np.sin(80*2*np.pi*x)
```

```python
yf = fft(y)
xf = np.linspace(0, 1./(2*dt), N//2)
```

```python
xf.shape, yf.shape
```

```python
plt.plot(xf, 2./N * np.abs(yf[:N//2]))
plt.grid()
# plt.xlim([0, 100])
plt.show()
```

---

My code for DFT:

```python
Y = np.zeros_like(y)
for kk in range(N):
    Yk=0.0
    for ii in range(N):
        Yk += y[ii] * np.exp(-2j*np.pi*kk*ii/N)
        Yk_abs = np.abs(Yk)
    Y[kk] = Yk_abs
```

```python
plt.plot(xf, 2./N * np.abs(Y[:N//2]))
plt.grid()
plt.show()
```
