Given $y = x^{-a}$<br>
Find $a$.

`````python
import numpy as np
import matplotlib.pyplot as plt
`````

`````python
x_List = np.linspace(0, 10, 101)
`````

`````python
x_List = x_List[1:]
`````

`````python
len(x_List)
`````

`````python
a = 2.3
`````

`````python
MyNoise = np.random.random(len(x_List))*0.005
y_List = x_List**-a + MyNoise
`````

`````python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([1.2, 0, 1, 1])

ax1.plot(x_List, y_List, "o")
ax2.plot(x_List, y_List, "o")

ax2.set_xscale("log")
ax2.set_yscale("log")

for ax in [ax1, ax2]:
    ax.set_xlabel("x", fontsize=20)
    ax.set_ylabel("y", fontsize=20)
    ax.grid(True)
plt.show()
`````

`````python
LogOfX = np.log(x_List)
LogOfY = np.log(y_List)
`````

`````python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(LogOfX, LogOfY, "o", alpha=0.5)

ax.set_xlabel("x", fontsize=20)
ax.set_ylabel("y", fontsize=20)

ax.grid(True)
plt.show()
`````

Fit a straight line to the above plot:

`````python
FittingResults = np.polyfit(LogOfX, LogOfY, 1, full=True)
FittingResults
`````

Here, <br>
Coefficients: ````[-2.20519029e+00, -1.43137945e-04]````<br>
Residuals: ```[1.23225578]```<br>
Rank: ```2```<br>
Singular values: ```[1.34996042, 0.4214343]```<br>
Conditioning threshold: ```2.2204460492503131e-14```

The residual is the sum of the squares of the fit errors.

`````python
MyCoeff = FittingResults[0]
FittedFn = np.poly1d(MyCoeff)
`````

`````python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(LogOfX, LogOfY, "o", alpha=0.5)
ax.plot(LogOfX, FittedFn(LogOfX), "-")

ax.text(1, 4, r"$a\approx${:.5f}".format(MyCoeff[0]), fontsize=20)

ax.set_xlabel("x", fontsize=20)
ax.set_ylabel("y", fontsize=20)

ax.grid(True)
plt.show()
`````
