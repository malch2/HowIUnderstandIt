Given $y = y_{0} e^{ax}$<br>
Find $a$.

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
x_List = np.linspace(0, 10, 21)
````

````python
x_List = x_List[1:]
````

````python
len(x_List)
````

````python
a = 2.3
y0 = 1e-2
````

````python
MyNoise = np.random.random(len(x_List))*10
y_List = y0 * np.exp(a*x_List) + MyNoise
````

````python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([1.2, 0, 1, 1])

ax1.plot(x_List, y_List, "o")
ax2.plot(x_List, y_List, "o")

# ax2.set_xscale("log")
ax2.set_yscale("log")

for ax in [ax1, ax2]:
    ax.set_xlabel("x", fontsize=20)
    ax.set_ylabel("y", fontsize=20)
    ax.grid(True)
plt.show()
````

````python
# LogOfX = np.log(x_List)
LogOfY = np.log(y_List)
````

````python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x_List, LogOfY, "o", alpha=0.5)

ax.set_xlabel("x", fontsize=20)
ax.set_ylabel("y", fontsize=20)

ax.grid(True)
plt.show()
````

Fit a straight line to the above plot:

````python
FittingResults = np.polyfit(x_List, LogOfY, 1, full=True)
FittingResults
````

Here, <br>
Coefficients: ```[1.67232033, 0.11210973]```<br>
Residuals: ```[32.60972754]```<br>
Rank: ```2```<br>
Singular values: ```[1.36986243, 0.35139285]```<br>
Conditioning threshold: ```4.440892098500626e-15```

The residual is the sum of the squares of the fit errors.

````python
MyCoeff = FittingResults[0]
FittedFn = np.poly1d(MyCoeff)
````

````python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x_List, LogOfY, "o", alpha=0.5)
ax.plot(x_List, FittedFn(x_List), "-")

ax.text(1, 4, r"$a\approx${:.5f}".format(MyCoeff[0]), fontsize=20)

ax.set_xlabel("x", fontsize=20)
ax.set_ylabel("y", fontsize=20)

ax.grid(True)
plt.show()
````
